```python
import pandas as pd
import numpy as np


# def get_grade(score):
#     grades = {
#         96 <= score <= 100: 'A',
#         90 <= score <= 95: 'A-',
#         87 <= score < 89: 'B+',
#         83 <= score < 86: 'B',
#         80 <= score < 82: 'B-',
#         77 <= score < 79: 'C+',
#         73 <= score < 76: 'C',
#         70 <= score < 72: 'C-',
#         67 <= score < 69: 'D+',
#         63 <= score < 66: 'D',
#         60 <= score < 62: 'D-',
#         score < 60: 'F'
#     }
#     return grades[True]


def get_grade(score):
    if 96 <= score <= 100:
        return 'A'
    elif 90 <= score <= 95:
        return 'A-'
    elif 87 <= score < 89:
        return 'B+'
    elif 83 <= score < 86:
        return 'B'
    elif 80 <= score < 82:
        return 'B-'
    elif 77 <= score < 79:
        return 'C+'
    elif 73 <= score < 76:
        return 'C'
    elif 70 <= score < 72:
        return 'D+'
    elif 67 <= score < 69:
        return 'D'
    elif 63 <= score < 66:
        return 'D-'
    elif 60 <= score < 62:
        return 'D'
    else:
        return 'F'
    


writer = pd.ExcelWriter("ur_dun.xlsx")

result = pd.read_excel("data.xlsx",sheet_name="Нийт")




irts_excel = pd.read_excel("data.xlsx",sheet_name="Ирц")
for row_index,row in irts_excel.iterrows():
    count = 0
    for each_col in row.items():
        if each_col[1] == "и":
            count +=1
    onoo = round((10 * count) / 32,2)
    irts_excel.loc[row_index,"Нийт -10"] = onoo
    result.loc[row_index,'Ирц-10'] = onoo

seminar_excel = pd.read_excel("data.xlsx",sheet_name="Семинар")
for row_index,row in seminar_excel.iterrows():
    count = 0
    for each_col in row.items():
        if each_col[1] == 1:
            count +=1
    onoo = round((25 * count) / 19,2)
    seminar_excel.loc[row_index,"Нийт -25"] = onoo
    result.loc[row_index,'Семинар-25'] = onoo


shalgalt_excel = pd.read_excel("data.xlsx",sheet_name="Шалгалт")
column_names = shalgalt_excel.columns.tolist()
last_column_index = len(column_names) - 1
sum_of_columns = shalgalt_excel[['Шалгалт - 1', 'Шалгалт - 2', 'Шалгалт - 3']].sum(axis=1)
shalgalt_excel.loc[:, column_names[last_column_index]] = sum_of_columns
result.loc[:,'Шалгалт-30'] = sum_of_columns




biy_daalt_excel = pd.read_excel("data.xlsx",sheet_name="Бие даалт")
column_names = biy_daalt_excel.columns.tolist()
last_column_index = len(column_names) - 1
sum_of_columns = biy_daalt_excel[['Бие даалт - 1', 'Бие даалт - 2']].sum(axis=1)
biy_daalt_excel.loc[:, column_names[last_column_index]] = sum_of_columns
result.loc[:,'Бие даалт-25'] = sum_of_columns



irts_excel.to_excel(writer, sheet_name="Ирц", index=False)
seminar_excel.to_excel(writer, sheet_name="Семинар", index=False)
shalgalt_excel.to_excel(writer, sheet_name="Шалгалт", index=False)
biy_daalt_excel.to_excel(writer, sheet_name="Бие даалт", index=False)


humuujil_excel = pd.read_excel("data.xlsx",sheet_name="Хүмүүжил хандлага")
column_names = humuujil_excel.columns.tolist()
last_column_index = len(column_names) - 1
sum_of_columns = humuujil_excel[['ХХ-1', 'ХХ-2','ХХ-3','ХХ-4']].sum(axis=1)
humuujil_excel.loc[:, column_names[last_column_index]] = sum_of_columns
result.loc[:,'ХХ-10'] = sum_of_columns




nemelt_excel = pd.read_excel("data.xlsx",sheet_name="Нэмэлт оноо")
column_names = nemelt_excel.columns.tolist()
last_column_index = len(column_names) - 1
sum_of_columns = nemelt_excel[['Нэмэлт оноо']].sum(axis=1)
nemelt_excel.loc[:, column_names[last_column_index]] = sum_of_columns
result.loc[:,'Нэмэлт оноо'] = sum_of_columns



irts_excel.to_excel(writer, sheet_name="Ирц", index=False)

seminar_excel.to_excel(writer, sheet_name="Семинар", index=False)
seminar_last = len(seminar_excel.columns.tolist())-1
shalgalt_excel.to_excel(writer, sheet_name="Шалгалт", index=False)
shalgalt_last = len(shalgalt_excel.columns.tolist())-1
humuujil_excel.to_excel(writer, sheet_name="Хүмүүжил хандлага", index=False)
humuujil_last = len(humuujil_excel.columns.tolist())-1
biy_daalt_excel.to_excel(writer, sheet_name="Бие даалт", index=False)
biydaalt_last = len(biy_daalt_excel.columns.tolist())-1
nemelt_excel.to_excel(writer, sheet_name="Нэмэлт оноо", index=False)
nemelt_last = len(nemelt_excel.columns.tolist())-1


result = result.rename(columns={'Unnamed: 11': 'Үсгэн дүн'})
result = result.rename(columns={'Unnamed: 13': 'Дүнгийн харьцаа'})

column_names_result = result.columns.tolist()
last_column_index = len(column_names_result) - 1
sum_of_columns = result[['Ирц-10','Семинар-25','Шалгалт-30','Бие даалт-25','ХХ-10','Нэмэлт оноо']].sum(axis=1)
result.loc[:, 'Нийт'] = sum_of_columns


result.loc[:,'Үсгэн дүн'] = sum_of_columns.apply(get_grade)
# result.loc[:,'Дүнгийн харьцаа'] = sum_of_columns.apply(get_count)



usgen_dun = result.loc[:,'Үсгэн дүн'].iloc[:-1]


grade_counts = usgen_dun.str.replace('[+-]', '').value_counts()

grade_counts = grade_counts.reindex(['A', 'B', 'C', 'D', 'F'])

amjilt = 0
chanar = 0



sum1 = 0
for key,value in grade_counts.items():
    if key != "F":
        sum1 += value

sum2 = 0
for key,value in grade_counts.items():
    if key == "A" or key == "B":
        sum2 += value
        

amjilt= str(round(sum1 / 28,2)) + "%"
chanar = str(round(sum2 / 28,2)) + "%"


too = []
sumToo = 0
for value in grade_counts:
    too.append(int(value))
    sumToo+=int(value)
too.append(sumToo)
customSeries = pd.Series(too, index=[0,1,2,3,4,5])
result.loc[:,'Дүнгийн харьцаа'] = customSeries



result.loc[12,'Unnamed: 12'] = 'Амжилт'
result.loc[13,'Unnamed: 12'] = 'Дүнгийн харьцаа'

result.loc[12,'Дүнгийн харьцаа'] = amjilt
result.loc[13,'Дүнгийн харьцаа'] = chanar
result.loc[13,'O'] = "Хамгийн их"
result.loc[14,'O'] = "Хамгийн бага"
result.loc[15,'O'] = "Дундаж"






# result.loc[12,'P'] = "Ирц-10"
# result.loc[13,'P'] = result.loc[:,'Ирц-10'].max()
# result.loc[14,'P'] = result.loc[:,'Ирц-10'].min()
# result.loc[15,'P'] = round(result.loc[:,'Ирц-10'].sum() / 28,2)


# result.loc[12,'Q'] = "Семинар-25"
# result.loc[13,'Q'] = result.loc[:,'Семинар-25'].max()
# result.loc[14,'Q'] = result.loc[:,'Семинар-25'].min()
# result.loc[15,'Q'] = round(result.loc[:,'Семинар-25'].sum() / 28,2)



# result.loc[12,'R'] = "Шалгалт-30"
# result.loc[13,'R'] = result.loc[:,'Шалгалт-30'].max()
# result.loc[14,'R'] = result.loc[:,'Шалгалт-30'].min()
# result.loc[15,'R'] = round(result.loc[:,'Шалгалт-30'].sum() / 28,2)

# result.loc[12,'S'] = "Бие даалт-25"
# result.loc[13,'S'] = result.loc[:,'Бие даалт-25'].max()
# result.loc[14,'S'] = result.loc[:,'Бие даалт-25'].min()
# result.loc[15,'S'] = round(result.loc[:,'Бие даалт-25'].sum() / 28,2)

# result.loc[12,'T'] = "ХХ-10"
# result.loc[13,'T'] = result.loc[:,'ХХ-10'].max()
# result.loc[14,'T'] = result.loc[:,'ХХ-10'].min()
# result.loc[15,'T'] = round(result.loc[:,'ХХ-10'].sum() / 28,2)

# result.loc[12,'U'] = "Нэмэлт оноо"
# result.loc[13,'U'] = result.loc[:,'Нэмэлт оноо'].max()
# result.loc[14,'U'] = result.loc[:,'Нэмэлт оноо'].min()
# result.loc[15,'U'] = round(result.loc[:,'Нэмэлт оноо'].sum() / 28,2)



columns = ['P', 'Q', 'R', 'S', 'T', 'U']
columnsName = ['Ирц-10', 'Семинар-25', 'Шалгалт-30', 'Бие даалт-25', 'ХХ-10', 'Нэмэлт оноо']
i=0
for col in columns:
    max_val = result.loc[:,columnsName[i]].max()
    min_val = result.loc[:,columnsName[i]].min()
    avg_val = round(result.loc[:,columnsName[i]].sum() / 28, 2)

    result.loc[12, col] = columnsName[i]
    result.loc[13, col] = max_val
    result.loc[14, col] = min_val
    result.loc[15, col] = avg_val
    i+=1




last_row = len(result.index) -1


dunguud = result.loc[:,'Нийт'].iloc[:-1].sum() / 28





result.loc[last_row,'Нийт'] = dunguud




# result.loc[:,'Дүнгийн харьцаа'] = x




result.to_excel(writer, sheet_name="Нийт", index=False)




    




writer.save()
writer.close()
```
